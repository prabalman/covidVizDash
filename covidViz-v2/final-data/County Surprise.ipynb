import bayesian_surprise3 as bs3

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

# County Case / Population

```python
f = './county-data/county_cases-population.csv'

dates, data = bs3.read_data(f, 5)
results = bs3.calc_surprise(data, ['base-rate', 'minimum', 'maximum', 'uniform'])

# data
# dates
# results['surprise']

ccp = pd.DataFrame(results['surprise']).T
ccp.columns = tuple(dates[1:])
ccp
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 8))

sns.lineplot(data=ccp.T, legend=False)

ax.set_title('County Case / Population')
ax.set_xlabel('Months')
ax.set_ylabel('Ratio')

ax.set_xticklabels(ccp.columns, rotation=45)
```

```python
ccp_history = pd.DataFrame(results['history']).T

# scp_history

fig, ax = plt.subplots(1,1, figsize=(10, 6))

sns.lineplot(data=ccp_history.T, legend=True, ax=ax)
ax.set_ylim(0.0, 1.0)
```

```python
# sanity check
print('sum: ', sum(results['pM'].values()))
print(results['pM'])

ccp.to_csv('./county-surprise/county_cases_population.csv')
```
