```python
import bayesian_surprise3 as bs3

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.set_theme('ggplot', context='notebook')
# sns.set_style('dark')
```

# Unemployment

```python
f = './unemployment/data.csv'

dates, data = bs3.read_data(f, None)
results = bs3.calc_surprise(data, ['base-rate', 'boom', 'bust'])
```

```python
unemployment = pd.DataFrame(results['surprise']).T
unemployment.columns = [str(year) for year in range(1981, 1999)]
# unemployment
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 8))

sns.lineplot(data=unemployment.T, legend=False)

ax.set_title('State Unemployment [Paper Example]')
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment')

ax.set_xticklabels(unemployment.columns, rotation=45)
```

```python
results['pM']
results['history']
u_history = pd.DataFrame(results['history']).T

# u_history

fig, ax = plt.subplots(1,1, figsize=(10, 6))

sns.lineplot(data=u_history.T, legend=True, ax=ax)

ax.set_ylim(0.0, 1.0)
```

# State Case / Population

```python
f = './state-data/state_case-population.csv'

dates, data = bs3.read_data(f, 2)
results = bs3.calc_surprise(data, ['base-rate', 'minimum', 'maximum', 'uniform'])

# data
# dates
# results['surprise']

scp = pd.DataFrame(results['surprise']).T
scp.columns = tuple(dates[1:])
# scp
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 8))

sns.lineplot(data=scp.T, legend=False)

ax.set_title('State Case / Population')
ax.set_xlabel('Months')
ax.set_ylabel('Ratio')

ax.set_xticklabels(scp.columns, rotation=45)
```

```python
scp_history = pd.DataFrame(results['history']).T

# scp_history

fig, ax = plt.subplots(1,1, figsize=(10, 6))

sns.lineplot(data=scp_history.T, legend=True, ax=ax)
ax.set_ylim(0.0, 1.0)
```

```python
# sanity check
print('sum: ', sum(results['pM'].values()))
print(results['pM'])

scp.to_csv('./state-surprise/state_cases_population.csv')
```
